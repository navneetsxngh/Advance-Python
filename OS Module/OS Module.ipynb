## OS Module

### In-Built Functions

```python
import os
```

```python
## os.path.exists() --> It checks if a path exists or not
os.path.exists('../hello.py')
```

```python
os.path.exists('../hello.py/')
```

```python
## os.path.isfile() --> 
os.path.isfile('../hello.py')
```

```python
## os.path.isdir() -->
os.path.isdir('../')
```

```python
os.path.isdir('../hello.py')
```

```python
## os.path.split() --> It splits the path and file name
os.path.split('../hello.py')
```

```python
## os.path.join() --> It joins the path and file name
os.path.join('..', 'hello.py')
```

```python
## os.path.basename() --> It gives the file name
os.path.basename('../hello.py')
```

```python
## os.path.dirname() --> It gives the directory name
os.path.dirname('../hello.py')
```
