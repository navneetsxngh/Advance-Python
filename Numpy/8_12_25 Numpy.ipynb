## NumPy

```python
import numpy as np
```

```python
l1 = [1.0,2,3,4,5,6]
arr = []

for i in l1:
    arr.append(i)
print(arr)
```

```python
arr1 = np.array([1,2,3])
print(type(arr))
```

```python
## 2d Array using the 2d list
arr2 = np.array([[1,2,3],[4,5,6]])
arr2
```

```python
## 3D Array
arr3 = np.array([[[1,2,3]]])
```

```python
## Creating Arrays with Zeros using [np.zeros()]
# syntax ---> np.zeros((rows, column))
arr0 = np.zeros((4,5)) 
arr0
arr0.ndim
```

```python
## Similarly [np.ones()]
# syntax ---> np.ones((rows,columns))

arr1 = np.ones((5,6), dtype= int)
arr1
```

```python
## [np.eye()]
# syntax ---> np.eye(order)

arri = np.eye(3, dtype= int)
arri
```

```python
## Creating Sequence with [np.arange()]
# syntax ---> np.arange(start, stop, step)

arr_arange = np.arange(1,10).reshape(3,3)
arr_arange
```

```python
## Creating evenly spaced numbers with np.linspace()
# syntax --> np.linspace(start, stop(inclusive), number of elements)
arr_lin = np.linspace(10,20,5)
arr_lin
```

```python
## Specifyiny Data Types (dtype) during Array Creation
arr = np.ones((3,4), dtype=str)
arr
```

```python
## Create a matrix (20,100) whose shape is its factors
m = np.arange(20,100).reshape(10,8)
m
```
